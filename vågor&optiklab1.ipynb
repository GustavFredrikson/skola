```
import matplotlib.pyplot as plt

x = [x*10 for x in range(-9, 10)]
print(x)
y = [0, 0.8, 2.7, 5.6, 8.9, 12.3, 15.4, 17.7, 19.2, 19.8, 19.3, 17.7, 15.3, 12.4, 8.8, 5.5, 2.7, 0.8, 0]

plt.plot(x,y)
plt.savefig('Experimentell.png')
```

```
import math

def intensity(theta):
    return 24 * math.cos(theta)**2

y_theoretic = []

for degree in x:
    y_theoretic.append(intensity(math.radians(degree)))

print(math.cos(math.radians(90)))
print(y_theoretic)
plt.plot(x,y_theoretic)
plt.savefig('Teoretisk.png')
```

```
for i in range(len(y_theoretic)):
    print(abs(y_theoretic[i] - y[i]))
```
